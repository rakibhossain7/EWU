# Python Overview & Data Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating Series
s = pd.Series([1,2,3,4,5], index=('a', 'b', 'c', 'd', 'e'))
s
```

```python
#DataFrame with Multi indexes
row_index = pd.MultiIndex.from_tuples([('Left', 'Football'), ('Right', 'Football'), ('Left', 'Cricket'), ('Right', 'Cricket')])
col_index = pd.MultiIndex.from_tuples([('Amateur', 'Male'), ('Amateur', 'Female'), ('Professional', 'Male'),('Professional', 'Female')])
data = [
    [5, 10, 15, 20], [3, 6, 9, 7], [2,6,4,8], [12, 14, 17, 19]
]
df = pd.DataFrame(data, index=row_index, columns=col_index)
df
```

```python
#Filtering
s_index = s.index
s_value = s.values
print("\n", s_index, s_value)
```

```python
#Functions
x_bar = s.mean()
x_min = s.min()
x_max = s.max()
x_std = s.std()
print("\n", 'mean:',x_bar, 'min:',x_min, 'max:',x_max, 'standar deviation:',x_std)
```

```python
#DataFrame using a list of Disctionaries
df = pd.DataFrame({'a': [1,3,5], 'b': [2,4,6], 'c': [3,6,9]}, index= ['p', 'q', 'r'])
df
```

```python
#Data Merging
stud_info = {'sub_id': ['1', '2', '3', '4', '5'], 'name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung']}
stud_info2 = {'sub_id': ['9', '10', '6', '7', '8'], 'name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty']}
stud_marks = {'sub_id': ['1', '2', '3', '4', '5'], 'marks': [91, 80, 75, 88, 90]}
stud_marks2 = {'sub_id': ['9', '10', '6', '7', '8'], 'marks': [81, 80, 75, 88, 90]}
batch301 = pd.DataFrame(stud_info, columns = ['sub_id', 'name'])
batch302 = pd.DataFrame(stud_info2, columns = ['sub_id', 'name'])
marks301 = pd.DataFrame(stud_marks, columns = ['sub_id', 'marks'])
marks302 = pd.DataFrame(stud_marks2, columns = ['sub_id', 'marks'])
pd.merge(batch301, marks301, on='sub_id')
```

```python
pd.concat([batch301, batch302])
```

```python
#Data Grouping
raw = {
    'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
    'subject_id': ['sub1', 'sub2', 'sub4', 'sub6', 'sub5'],
    'Marks_scored': [98, 90, 87, 69, 78],
    'Year': [2009, 2010, 2011, 2009, 2010]
}
df = pd.DataFrame(raw)
group = df.groupby('Year')
print(df)
print(group.get_group(2009))
```
